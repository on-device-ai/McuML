```python
import os

root_path = ''
proj_path = ''
class cd:
    """Context manager for changing the current working directory"""
    def __init__(self, newPath):
        self.newPath = os.path.expanduser(newPath)

    def __enter__(self):
        self.savedPath = os.getcwd()
        os.chdir(self.newPath)

    def __exit__(self, etype, value, traceback):
        os.chdir(self.savedPath)
        
with cd('..\\..\\'):
    root_path = os.getcwd()
proj_path = os.getcwd()
```

# Deploy to STM32F746

Since Mbed requires a special folder structure for projects, we'll first run a command to generate a subfolder containing the required source files in this structure:

```python
os.chdir(root_path + '\\tensorflow')
```

Issue : https://github.com/tensorflow/tensorflow/issues/43201

```python
cmd = 'bash -c "make -f tensorflow/lite/micro/tools/make/Makefile TARGET=mbed TAGS=\\"CMSIS disco_f746ng\\" generate_micro_speech_mbed_project"'
print(os.popen(cmd).read())
```

This will result in the creation of a new folder:

```python
os.chdir('tensorflow\\lite\\micro\\tools\\make\\gen\\mbed_cortex-m4\\prj\\micro_speech\\mbed')
```

First, tell Mbed that the current directory is the root of an Mbed project:

```python
cmd = proj_path + '\\mbed_cmd.bat mbed config root .'
print(os.popen(cmd).read())
```

Next, tell Mbed to download the dependencies and prepare to build:

```python
cmd = proj_path + '\\mbed_cmd.bat mbed deploy'
print(os.popen(cmd).read())
```

By default, Mbed will build the project using C++98. However, TensorFlow Lite requires C++11. Run the following Python snippet to modify the Mbed configuration files so that it uses C++11:

```python
import fileinput, glob;
for filename in glob.glob("mbed-os/tools/profiles/*.json"):
    for line in fileinput.input(filename, inplace=True):
        print(line.replace("\"-std=gnu++98\"","\"-std=c++11\", \"-fpermissive\""))
```

Finally, run the following command to compile:

```python
cmd = proj_path + '\\mbed_cmd.bat mbed compile -m DISCO_F746NG -t GCC_ARM'
print(os.popen(cmd).read())
```

To deploy, plug in your STM board and copy the file to it. You can do this with the following command:

```python
DIS_F746NG = 'D:\\'
cmd = 'copy .\\BUILD\\DISCO_F746NG\\GCC_ARM\\mbed.bin ' + DIS_F746NG
print(os.popen(cmd).read())
```
